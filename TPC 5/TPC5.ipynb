```python
# TPC 5

cinema1 = []

def listar (cinema):
    for sala in cinema:
        nlugares,vendidos,filme = sala
        print ("Filme:",filme)

def disponivel (cinema,nome,lugar):
    cond = False
    for sala in cinema:
        nlugares,vendidos,filme = sala
        if nome == filme and lugar <= nlugares:
            if lugar not in vendidos:
                cond = True 
    return cond

def vendebilhete (cinema,nome,lugar):
    for sala in cinema:
        nlugares,vendidos,filme = sala
        if nome == filme and lugar <= nlugares:
            vendidos.append((lugar))
    return cinema

def listardisponibilidades (cinema):
    for sala in cinema:
        nlugares,vendidos,filme = sala
        if len(vendidos) == nlugares:
            print ("Os lugares estão todos vendidos!")
        else:
            print ("Filme:",filme,"--","Lugares disponíveis:",(nlugares-len(vendidos)))

def inserirSala (cinema,sala):
    if sala not in cinema:
        cinema.append(sala)
    return cinema

def menu():
    print("""
    (1) Consultar todos os filmes em exibição no cinema
    (2) Verificar se um lugar está disponível
    (3) Vender um bilhete
    (4) Consultar quantos lugares ainda estão disponíveis na sala
    (5) Acrescentar uma nova sala
    (0) Sair """)

sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)

menu()
opcao = int(input("Introduza uma opção: "))

while opcao != 0:

    if opcao == 1:
        print(listar(cinema1))

    if opcao == 2:
        if disponivel (cinema1,input("Insira o nome do filme:"),int(input("Insere o número do lugar:"))) == True:
            print ("O bilhete está disponível!")
        else:
            print ("O bilhete não está disponível!")

    if opcao == 3:
        print(vendebilhete (cinema1,input("Insira o nome do filme"), int(input("Insira o número do lugar:"))))

    if opcao == 4:
        print(listardisponibilidades (cinema1))

    if opcao == 5:
        sala=[]
        sala.append(int(input("Insira o número de lugares da sala:")))
        sala.append([])
        sala.append(str(input("Insira o nome do filme em exibição:")))
        print(inserirSala (cinema1,sala))
        
    opcao = int(input("O que pretende consultar de seguida?"))
```
