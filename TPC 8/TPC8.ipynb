```python
# quantosPost, que indica quantos posts estão registados:

def quantosPost(redeSocial): 
    return len(redeSocial)

# postsAutor, que devolve a lista de posts de um determinado autor:

def postsAutor(redeSocial, autor):
    for dicionario in redeSocial:
        for chave in dicionario:
            if dicionario[chave] == autor:
                res = dicionario["conteudo"]
    return res

# autores, que devolve a lista de autores de posts ordenada alfabeticamente:

def autores(redeSocial):
    res = []
    for dicionario in redeSocial:
        j = dicionario["autor"]
        res.append(j)
    return res

# insPost, que acrescenta um novo post à rede social a partir dos parâmetros recebidos e devolve a nova rede social:

def insPost(redeSocial, conteudo, autor, dataCriacao, comentarios):
    i = len(redeSocial) + 1
    j = "p" + str(i)
    redeSocial.append({"id":j,"conteudo":conteudo,"autor":autor,"dataCriacao":dataCriacao,"comentarios":comentarios})
    return redeSocial

# remPost, que remove um post da rede, correspondente ao `id` recebido:

def remPost(redeSocial, id):
    for dicionario in redeSocial:
        if dicionario["id"] == id:
            redeSocial.remove(dicionario)
    return 

# postsPorAutor, que devolve uma distribuição de posts por autor (à semelhança do que foi feito nas aulas):

def postsPorAutor(redeSocial):
    distrib = {}
    for dicionario in redeSocial:
        autor = dicionario['autor']
        if autor in distrib:
            distrib[autor] = distrib[autor] + 1
        else:
            distrib[autor] = 1
    return distrib

# comentadoPor, que recebe um autor e devolve a lista de posts comentados por esse autor:

def comentadoPor(redeSocial, autor):
    lista = []
    for dicionario in redeSocial:
        for comentario in dicionario['comentarios']:
            if comentario['autor'] == autor:
                lista.append(dicionario)
    return lista
```
