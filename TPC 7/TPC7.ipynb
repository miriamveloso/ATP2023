    2. Define uma função que recebe dois polinómios `p1` e `p2` e calcula um novo polinómio correspondente à soma de `p1` e `p2`;

* Num ficheiro Python, cria uma aplicação para manipular polinómios com as seguintes operações:
    1. Criar um polinómio interativamente;
    2. Ler uma lista de polinómios de um ficheiro;
    3. Listar polinómios: mostra numa tabela os polinómios carregados em memória, adicionando um número de ordem a cada um;
    4. Calcular o valor de um polinómio num ponto (usa o número de ordem para referenciar o polinómio);
    5. Listar polinómios com grau: à listagem anterior, acrescenta mais uma coluna com o grau;
    6. Maior grau: mostra o polinómio de maior grau e indica o seu número de ordem;
    7. Derivada: mostra uma tabela com os polinómios e as respetivas derivadas;
    8. Somar dois polinómios: indicando os seus números de ordem;
    9. Gerar um gráfico para o polinómio;
    10. Gravar num ficheiro os polinómios em memória;
    0. Sair da aplicação


```python
# TPC 7
# Instancia os seguintes exemplos num ficheiro de texto com o formato definido na última aula

pol1 = [(1,7),(-3.7,4),(5,3),(-88,0)]
pol2 = [(2,2),(17,0)]
pol3 = [(7,4),(6,3),(-1,2),(43,0)]
pol4 = [(1,4)]
pol5 = [(1,2)]
pol6 = [(1,3)]

def menu ():
    print (''' Menu:
    (1) Criar um polinómio interativamente;
    (2) Ler uma lista de polinómios de um ficheiro;
    (3) Listar polinómios: mostra numa tabela os polinómios carregados em memória, adicionando um número de ordem a cada um;
    (4) Calcular o valor de um polinómio num ponto (usa o número de ordem para referenciar o polinómio);
    (5) Listar polinómios com grau: à listagem anterior, acrescenta mais uma coluna com o grau;
    (6) Maior grau: mostra o polinómio de maior grau e indica o seu número de ordem;
    (7) Derivada: mostra uma tabela com os polinómios e as respetivas derivadas;
    (8) Somar dois polinómios: indicando os seus números de ordem;
    (9) Gerar um gráfico para o polinómio;
    (10) Gravar num ficheiro os polinómios em memória;
    (11) Sair da aplicação
''')
    return

def guardaPolinomios (listap,fnome):
    fpol = open(fnome,"w")
    for pol in listap:
        for monomio in pol:
            coef,exp = monomio
            fpol.write(str(coef) + "," + str(exp) + "|" )
        fpol.write("\n")
    fpol.close()
    return 

# Cria um polinómio iterativamente

def criaPol ():
    lista = []
    exp = int(input("Digite o grau do polinómio:"))
    while exp >= 0:
        coef = int(input("Digite o coeficiente de grau " + str(exp) + ":"))
        if coef != 0:
            lista.append((coef,exp))
        exp = exp - 1
    return lista

# Ler uma lista de polinómios de um ficheiro

def recuperarPolinomios(fnome):
    listaPol = []
    f1 = open(fnome)
    for linha in f1: #percorrer as linhas do ficheiro
        pol = [] #lista de cada polinómio
        monomios = linha.split("|") #vamos isolar os monómios que estão separados por |
        for mx in monomios: # o m é texto, temos que separar das virgulas
            m = mx.split(',')
            if len(m) == 2:
                pol.append(((float(m[0])),int(m[1]))) # selecionamos os dois elementos de cada monómio
        listaPol.append(pol)
    f1.close()
    return listaPol

# Listar polinómios: mostra numa tabela os polinómios carregados em memória, adicionando um número de ordem a cada um

def listarPol(fnome,linhas):
    f = open(fnome)
    i = 1
    for linhas in f:
        print(str(i) + " :: " + linhas, end="")
        i = i+1
    return

# Calcular o valor de um polinómio num ponto (usa o número de ordem para referenciar o polinómio)

def calcPolinomio(p, x):
    res = 0
    for i in p:
        coef, grau = i
        res = coef * (x**grau)+res
    return res

# Listar polinómios com grau: à listagem anterior, acrescenta mais uma coluna com o grau

def listarGrau(fnome,linhas):
    f = open(fnome)
    i = 1
    for linhas in f:
        for j in f :
            grau = j
            print(str(i) + " :: " + linhas, end="" + " :: " + str(grau))
        i = i+1
    return


# # Maior grau: mostra o polinómio de maior grau e indica o seu número de ordem

def grauPolinomios(listap):
    maior = 0
    for p in listap:
        grau = calcPolinomio(p)
        if grau > maior:
            maior = grau
    return maior

# Define uma função que dado um polinómio calcula a sua derivada

def derivada(pol): 
    deriv = []
    for monomio in pol:
        coef,exp = monomio
        if exp != 0:
            deriv.append((coef*exp,exp-1))
    return deriv

derivada(pol1)

# Somar dois polinómios: indicando os seus números de ordem

def somarPolinomios(primeiroPol, segundoPol):
    somaPoL = []
    for x in primeiroPol:
        coe, grau = x
        encontrado = False
        for i in segundoPol:
            c, exp = i
            if grau == exp:
                somaPoL.append((c + coe, grau))
                encontrado = True
                segundoPol.remove(i)  
        if not encontrado:
            somaPoL.append((coe, grau))
    somaPoL = somaPoL + segundoPol  
    return somaPoL

# Gerar um gráfico para o polinómio;

import matplotlib.pyplot as plt

def graficoPol(p):
    x = list(range(-10,10))
    y = [sum(coef * (xi ** grau) for coef, grau in p) for xi in x]

    plt.figure(figsize=(5,5))
    plt.plot(x, y, label='Polinómio', color='b')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico do Polinómio')

    plt.grid()
    plt.show()

    
# Gravar num ficheiro os polinómios em memória;

def guardaPolinomios (listap,fnome):
    fpol = open(fnome,"w")
    for pol in listap:
        for monomio in pol:
            coef,exp = monomio
            fpol.write(str(coef) + "," + str(exp) + "|" )
        fpol.write("\n")
    fpol.close()
    return 


menu()
opção=int(input("Qual a opção que deseja selecionar?"))


polinomios = []

while opção != 0:
    if opção == 1:
        novo_polinomio = criaPol()
        polinomios.append(novo_polinomio)
        
    if opção == 2:
        nome_arquivo = input("Insira o nome do arquivo: ")
        recuperarPolinomios(nome_arquivo)
        
    if opção == 3:
        print(" Segue a tabela com a listagem dos polinomios:")
        nome_arquivo = input("Insira o nome do arquivo: ")
        listarPol(nome_arquivo, len(nome_arquivo))
        
    if opção == 4:
        x = int(input("Indique o valor de x para o cálculo do polinómio: "))
        valor = calcPolinomio(novo_polinomio, x)
        print(valor)
        
    if opção == 5:
            print(" Segue a tabela com a listagem dos polinomios com o valor de grau:")
            nome_arquivo = input("Insira o nome do arquivo: ")
            listarGrau(nome_arquivo, len(polinomios))
            print("")
    if opção == 6:
        print("O maior grau da lista é o", grauPolinomios(polinomios),".")
       
    if opção == 7:
        print("A derivada do polinómio é ", derivada(polinomios),".")
        
    if opção == 8:
        segundo = criaPol()
        print("A soma dos dois polinomios é ", somarPolinomios(polinomios, segundo),".")
        
    if opção == 9:
        graficoPol(polinomios)
       
    if opção == 10:
        nome = input("Digite o nome do arquivo para salvar os polinômios: ")
        guardaPolinomios(polinomios, nome)
        

    opção = int(input("O que pretende selecionar de seguida?"))

```
